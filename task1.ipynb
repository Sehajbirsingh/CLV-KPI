```python
import pandas as pd
import xlrd

# Path to the Excel workbook
excel_file = r"C:\Users\mynam\Downloads\Sample - Superstore(1) (1).xls"

# Read all sheets from the Excel file
xls = pd.ExcelFile(excel_file, engine='xlrd')

# Get the sheet names
sheet_names = xls.sheet_names

# Create a dictionary to store dataframes
dfs = {}

# Read each sheet into a separate dataframe
for sheet in sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet, engine='xlrd')
    dfs[sheet] = df

# Print the names of the sheets and the first few rows of each dataframe
for sheet_name, df in dfs.items():
    print(f"\nSheet: {sheet_name}")
    print(df.head())
    print("\n" + "="*50 + "\n")

# You can now access each dataframe using the sheet name as the key
# For example, to access the 'Orders' sheet dataframe:
# orders_df = dfs['Orders']
```

```python
# Assuming you've already run the code to load the sheets into dfs

# Print the names of all available sheets
print("Available sheets:")
for sheet_name in dfs.keys():
    print(f"- {sheet_name}")

# Access individual dataframes
orders_df = dfs['Orders']
returns_df = dfs['Returns']
users_df = dfs['People']

# Now you can work with each dataframe separately
print("\nFirst few rows of Orders:")
print(orders_df.head())

print("\nFirst few rows of Returns:")
print(returns_df.head())

print("\nFirst few rows of Users:")
print(users_df.head())

# You can also perform operations on individual dataframes
print("\nShape of Orders dataframe:", orders_df.shape)
print("Columns in Returns dataframe:", returns_df.columns.tolist())
print("Summary statistics of Users dataframe:")
print(users_df.describe())
```

```python
# Check the number of missing values in each column of orders_df
missing_values = orders_df.isnull().sum()

# Calculate the percentage of missing values
total_rows = len(orders_df)
missing_percentage = (missing_values / total_rows) * 100

# Combine the count and percentage into a single dataframe
missing_data = pd.concat([missing_values, missing_percentage], axis=1, keys=['Missing Count', 'Missing Percentage'])

# Sort the dataframe by the number of missing values in descending order
missing_data = missing_data.sort_values('Missing Count', ascending=False)

# Display the results
print("Missing values in orders_df:")
print(missing_data)

# Optionally, display only columns with missing values
print("\nColumns with missing values:")
print(missing_data[missing_data['Missing Count'] > 0])
```

```python

# Replace missing values in 'Postal Code' with '0000'
orders_df['Postal Code'] = orders_df['Postal Code'].fillna('0000')

# Verify the changes

print(orders_df['Postal Code'].isna().sum(), "missing values")





```

```python
#total number of rows in the dataframe
print(len(orders_df))

distinct_combinations = orders_df.groupby(['Order ID', 'Product ID']).size().reset_index().shape[0]
print(distinct_combinations)

```

```python
# Group by 'Order ID' and 'Product ID', count occurrences, and filter for counts > 1
duplicates = orders_df.groupby(['Order ID', 'Product ID']).size().reset_index(name='count')
duplicates = duplicates[duplicates['count'] > 1]

# Sort by count in descending order (optional, for better readability)
duplicates = duplicates.sort_values('count', ascending=False)

print("Duplicate combinations of 'Order ID' and 'Product ID':")
print(duplicates)

# Print the total number of duplicate combinations
print(f"\nTotal number of duplicate combinations: {len(duplicates)}")
```

```python
duplicates = orders_df.groupby(['Order ID', 'Product ID']).size().reset_index(name='count')
duplicates = duplicates[duplicates['count'] > 1]

# Now, merge this back with the original dataframe to get the desired columns
result = orders_df.merge(duplicates[['Order ID', 'Product ID']], on=['Order ID', 'Product ID'])

# Select only the columns we want
result = result[['Product ID', 'Order ID', 'Sales', 'Quantity', 'Discount', 'Profit']]

# Sort the result (optional, for better readability)
result = result.sort_values(['Order ID', 'Product ID'])

result

```

```python
merged_orders = orders_df.groupby(['Order ID', 'Product ID']).agg({
    'Order Date': 'max',
    'Ship Date': 'max',
    'Ship Mode': 'max',
    'Customer ID': 'max',
    'Customer Name': 'max',
    'Segment': 'max',
    'Country/Region': 'max',
    'City': 'max',
    'State': 'max',
    'Postal Code': 'max',
    'Region': 'max',
    'Category': 'max',
    'Sub-Category': 'max',
    'Product Name': 'max',
    'Sales': 'sum',
    'Quantity': 'sum',
    'Discount': 'max',
    'Profit': 'sum'
}).reset_index()
```

```python
merged_orders
```

```python
duplicates = merged_orders.groupby(['Order ID', 'Product ID']).size().reset_index(name='count')
duplicates = duplicates[duplicates['count'] > 1]

# Now, merge this back with the original dataframe to get the desired columns
result = merged_orders.merge(duplicates[['Order ID', 'Product ID']], on=['Order ID', 'Product ID'])

# Select only the columns we want
result = result[['Product ID', 'Order ID', 'Sales', 'Quantity', 'Discount', 'Profit']]

# Sort the result (optional, for better readability)
result = result.sort_values(['Order ID', 'Product ID'])

result
```
